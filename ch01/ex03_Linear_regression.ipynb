- Data definition

    - training dataset 
    - test dataset 
    - data는 torch.tensor를 이용해서 저장 
    - 입력(x_train)과 출력(y_train)은 각각 따로 저장한다. 
    - 입 출력은 x, y로 구분한다. 

|Hours(x)|Point(y)|
|:---:|:---:|
|Train dataset(x)|Train dataset(y)|
|1|2|
|2|4|
|3|6|
|Test dataset(x)|Test dataset(y)|
|4|?|
    

- Hypothesis 
$$y=Wx + b  \\\\ w=Weight \\\\ b = Bias $$




```python
import torch
import numpy as np 
import pandas as pd 
import os 
import sys 
import torch.optim as optim 
import torch.nn as nn 
import torch.nn.functional as F 
```

```python
torch.manual_seed(1)
```

```python
x_train = torch.FloatTensor([[1], [2], [3]])
y_train = torch.FloatTensor([[2], [4], [6]])
```

```python
w = torch.zeros(1, requires_grad=True)
b = torch.zeros(1, requires_grad=True)
hypothesis = x_train * w + b 
print(type(hypothesis))
```

- Compute loss 
    - Mean Squared Error(MSE)
$$cost(W,b ) = \frac{1}{m}\sum_\text{i=1}^m(H(x^\text{(i)}) - y^\text{(i)})^2$$ 



```python
cost = torch.mean((hypothesis  - y_train) **2)
print(cost)
```

- Gradient descent

```python
optimizer = optim.SGD([w, b], lr = 0.01)

optimizer.zero_grad()
cost.backward()
optimizer.step()
```

```python
print(w)
print(b)
```

```python
nb_epochs = 1000
```

```python
for epoch in range(1, nb_epochs+1):
    hypothesis = x_train * w +b 
    cost =torch.mean((hypothesis - y_train)**2)

    optimizer.zero_grad()
    cost.backward()
    optimizer.step()
    if epoch % 100 == 0:
        print('Epoch {:4d}/{} W: {:.3f}, b: {:.3f} Cost: {:.6f}'.format(
            epoch, nb_epochs, w.item(), b.item(), cost.item()
        ))
```
